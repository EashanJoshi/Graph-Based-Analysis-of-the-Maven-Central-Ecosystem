```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"D:\Neo4j Desktop\relate-data\dbmss\dbms-335ec2d2-42b4-48e3-bfe8-2ae904cf2831\import\all_results_2006_detailed.csv"  # Updated file path
data = pd.read_csv(file_path)

# Visualization 1: Distribution of Added and Removed Dependencies
def plot_dependency_distribution(data):
    """
    Plot the distribution of added and removed dependencies.
    """
    plt.figure(figsize=(12, 6))
    plt.hist(data['AddedDepsCount'], bins=30, alpha=0.7, label="Added Dependencies", color='green')
    plt.hist(data['RemovedDepsCount'], bins=30, alpha=0.7, label="Removed Dependencies", color='red')
    plt.title("Distribution of Added and Removed Dependencies", fontsize=16)
    plt.xlabel("Number of Dependencies", fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Visualization 2: Net Change Distribution
def plot_net_change_distribution(data):
    """
    Plot the distribution of net dependency changes.
    """
    plt.figure(figsize=(12, 6))
    plt.hist(data['NetChange'], bins=30, alpha=0.7, color='blue')
    plt.title("Distribution of Net Dependency Changes", fontsize=16)
    plt.xlabel("Net Change in Dependencies", fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Visualization 3: Change Type Summary
def plot_change_summary(data):
    """
    Plot the count of change types (Changed vs No Change).
    """
    change_summary = data['Change'].value_counts()
    plt.figure(figsize=(8, 6))
    change_summary.plot(kind='bar', color=['blue', 'orange'], alpha=0.7)
    plt.title("Change Summary (Changed vs No Change)", fontsize=16)
    plt.xlabel("Change Type", fontsize=14)
    plt.ylabel("Count", fontsize=14)
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Run the visualizations
plot_dependency_distribution(data)
plot_net_change_distribution(data)
plot_change_summary(data)
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Directory containing yearly CSV files
directory_path = r"D:\Neo4j Desktop\relate-data\dbmss\dbms-335ec2d2-42b4-48e3-bfe8-2ae904cf2831\import"

# Initialize an empty DataFrame to store results for all years
all_years_data = []

# Process each file in the directory
for year in range(2005, 2025):  # From 2006 to 2024
    file_name = f"all_results_{year}_detailed.csv"  # File naming pattern
    file_path = os.path.join(directory_path, file_name)
    
    # Check if the file exists
    if os.path.exists(file_path):
        print(f"Processing {file_name}...")
        # Load the CSV file
        data = pd.read_csv(file_path)
        
        # Aggregate total AddedDepsCount and RemovedDepsCount
        total_added = data['AddedDepsCount'].sum()
        total_removed = data['RemovedDepsCount'].sum()
        
        # Calculate the average NetChange per library
        average_net_change = data['NetChange'].mean()
        
        # Append the results to the all_years_data list
        all_years_data.append({
            "Year": year,
            "TotalAddedDeps": total_added,
            "TotalRemovedDeps": total_removed,
            "AverageNetChange": average_net_change
        })
    else:
        print(f"File not found: {file_name}")

# Convert the results into a DataFrame
aggregated_data = pd.DataFrame(all_years_data)

# Save the aggregated results to a CSV file (optional)
output_path = os.path.join(directory_path, "aggregated_results_2006_to_2024.csv")
aggregated_data.to_csv(output_path, index=False)
print(f"Aggregated results saved to {output_path}")

# Visualization: Total Added and Removed Dependencies by Year
plt.figure(figsize=(12, 6))
plt.plot(aggregated_data['Year'], aggregated_data['TotalAddedDeps'], label='Total Added Dependencies', marker='o', color='green')
plt.plot(aggregated_data['Year'], aggregated_data['TotalRemovedDeps'], label='Total Removed Dependencies', marker='o', color='red')
plt.title("RQ1:Total Added and Removed Dependencies (2005-2024)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Count of Dependencies", fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualization: Average Net Change by Year
plt.figure(figsize=(12, 6))
plt.plot(aggregated_data['Year'], aggregated_data['AverageNetChange'], label='Average Net Change', marker='o', color='blue')
plt.axhline(0, color='gray', linestyle='--', label="No Change")  # Reference line
plt.title("RQ2: Average Net Change in Dependencies (2005-2024)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Net Change", fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# Path to the directory containing CSV files for all years
directory_path = r"D:\Neo4j Desktop\relate-data\dbmss\dbms-335ec2d2-42b4-48e3-bfe8-2ae904cf2831\import"

# Lists to store aggregated results for all years
years = []
total_added_list = []
total_removed_list = []
average_net_change_list = []

# Aggregate data for all years
for year in range(2005, 2025):  # Iterate from 2005 to 2024
    file_path = os.path.join(directory_path, f"all_results_{year}_detailed.csv")
    
    if os.path.exists(file_path):
        # Load the data for the year
        data = pd.read_csv(file_path)

        # Aggregate data
        total_added = data['AddedDepsCount'].sum()
        total_removed = data['RemovedDepsCount'].sum()
        average_net_change = data['NetChange'].mean()

        # Append results to lists
        years.append(year)
        total_added_list.append(total_added)
        total_removed_list.append(total_removed)
        average_net_change_list.append(average_net_change)
    else:
        print(f"File not found for year {year}: {file_path}")

# Bar Chart 1: Total Added and Removed Dependencies Across Years
plt.figure(figsize=(12, 6))
bar_width = 0.4  # Width of each bar
x = range(len(years))  # Positions for the bars
plt.bar(x, total_added_list, width=bar_width, label="Total Added Dependencies", color="green", align="center")
plt.bar([p + bar_width for p in x], total_removed_list, width=bar_width, label="Total Removed Dependencies", color="red", align="center")
plt.title("Total Added and Removed Dependencies (20052024)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Count of Dependencies", fontsize=14)
plt.xticks([p + bar_width / 2 for p in x], years, rotation=45)  # Center ticks between bars
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Bar Chart 2: Average Net Change in Dependencies Across Years
plt.figure(figsize=(12, 6))
plt.bar(years, average_net_change_list, color="blue", align="center")
plt.axhline(0, color='gray', linestyle='--', label="No Change")  # Reference line
plt.title("Average Net Change in Dependencies (20052024)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Net Change", fontsize=14)
plt.xticks(years, rotation=45)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
